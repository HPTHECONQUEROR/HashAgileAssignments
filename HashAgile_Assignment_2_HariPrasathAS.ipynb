```python
import pandas as pd
from elasticsearch import Elasticsearch
```

## Creating collection with the name "employee_data"

```python
es = Elasticsearch([{'host': 'localhost', 'port': 9200, 'scheme': 'http'}])

def createCollection(p_collection_name):
    if not es.indices.exists(index=p_collection_name):
        es.indices.create(index=p_collection_name)
        print(f"Collection '{p_collection_name}' created.")
    else:
        print(f"Collection '{p_collection_name}' already exists.")

createCollection('employee_data')  
```

## Data cleaning (This is because the data may contain some unhashable special characters) (not mentioned on the assignment)

```python
def dataCleaning():
    data = pd.read_csv('emp.csv', encoding='ISO-8859-1')

    data['Annual Salary'] = data['Annual Salary'].replace({'\$': '', ',': '', ' ': ''}, regex=True).astype(float)
    data['Exit Date'] = data['Exit Date'].where(data['Exit Date'].notna(), None) 
    data.to_csv('cleaned_emp.csv', index=False)

    print("Data cleaning complete. Cleaned data saved to 'cleaned_emp.csv'.")

dataCleaning()
```

## Index Data - Indexing the data to specified column

```python
def indexData(p_collection_name, p_exclude_column):
    data = pd.read_csv('cleaned_emp.csv', encoding='ISO-8859-1')
    
    for _, row in data.iterrows():
        doc = row.drop(p_exclude_column).to_dict()
        try:
            es.index(index=p_collection_name, body=doc)
            print(f"Document indexed: {doc}")  
        except Exception as e:
            print(f"Error indexing data: {e} for document: {doc}")

indexData('employee_data', 'Exit Date')
```

## Searching using the Column records

```python

def searchByColumn(collection_name, column_name, column_value):
    search_query = {
        "query": {
            "match": {
                column_name: column_value
            }
        }
    }
    response = es.search(index=collection_name, body=search_query)
    return response['hits']['hits']
searchByColumn("employee_data", 'Department', 'IT')
```

## Getting the counts of the collection "employee_date"

```python
def getEmpCount(p_collection_name):
    count = es.count(index=p_collection_name)['count']
    print(f"Employee count in '{p_collection_name}': {count}")
    return count

getEmpCount('employee_data')
```

## Deleting by EmployeeID

```python
from elasticsearch.exceptions import NotFoundError

def delEmpById(collection_name, employee_id):
    try:
        es.delete(index=collection_name, id=employee_id)
        print(f"Employee with ID {employee_id} deleted from collection '{collection_name}'.")
    except NotFoundError:
        print(f"Employee with ID {employee_id} not found in collection '{collection_name}'.")
delEmpById("employee_data", 'E02004')
```

## Retrieving the count of employees grouped by department from the specified collection.

```python

def getDepFacet(collection_name):
    aggs = {
        "department_counts": {
            "terms": {
                "field": "Department.keyword",  # Use .keyword for aggregations
                "size": 10  # Limit to top 10 departments
            }
        }
    }
    response = es.search(index=collection_name, body={"aggs": aggs, "size": 0})
    return response['aggregations']['department_counts']['buckets']
getDepFacet("employee_data")
```
